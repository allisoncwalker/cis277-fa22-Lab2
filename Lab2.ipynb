```python
df.columns
```

```python
display(df.rating.unique())
display(df.votes.unique())
```

#### This code takes the * out of the number to make it an integer instead of a float.

```python
df.rating.replace("*9.0", 9.0, inplace=True)
display(df.rating.unique())
```

#### This code takes all the , out of Votes

```python
df['votes'] = df['votes'].str.replace(",", "")
display(df.votes.unique())
```

#### This code will turn all the strings in Ratings and Votes into integers.

```python
# df.Rating.astype(float)
# df.info()

df = df.astype({"rating": float, "votes": float})
df.info()

```
